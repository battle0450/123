```python
import pandas as pd
import numpy as np
import string
import time
```

```python
#讀檔
df = pd.read_excel("107年新竹站_20190315.xls")
```

```python
#將10/01以後的取出
df = df.loc[4908:,:]
```

```python
df
```

```python
df.info()
```

```python
#從0開始編index
df.index = range(0,len(df))
```

```python
#將所有NR替換成0
df = df.replace('NR',0)
```

```python
#先將非數值的屬性分離
df_item = df[['日期','測站','測項']]
df = df.drop(['日期','測站','測項'],axis = 1)
```

```python
#從左到右從上到下依序判斷是否有包含無效值和缺失值,不斷判斷到直到有合法值然後取平均
start = time.time()
for i in range(len(df)):
    for j in range(0,24):
            x = i
            y = j
            if str.find(str(df.iloc[i,j]),'#')!=-1 or str.find(str(df.iloc[i,j]),'*')!=-1 or str.find(str(df.iloc[i,j]),'x')!=-1 or str.find(str(df.iloc[i,j]),'A')!=-1 or np.isnan(df.iloc[i,j])==True:
                if j!=0 and j!=23: #居於1~22
                    if str.find(str(df.iloc[i,j+1]),'#')==-1 and str.find(str(df.iloc[i,j+1]),'*')==-1 and str.find(str(df.iloc[i,j+1]),'x')==-1 and str.find(str(df.iloc[i,j+1]),'A')==-1 and np.isnan(df.iloc[i,j+1])==False: #不包含
                        df.iloc[i,j] = (float(df.iloc[i,j-1])+float(df.iloc[i,j+1]))/2
                    else:
                        while str.find(str(df.iloc[x,y]),'#')!=-1 or str.find(str(df.iloc[x,y]),'*')!=-1 or str.find(str(df.iloc[x,y]),'x')!=-1 or str.find(str(df.iloc[x,y]),'A')!=-1 or np.isnan(df.iloc[x,y])==True:
                            y+=1
                            if y==24:
                                x+=18
                                y = 0
                        df.iloc[i,j] = (float(df.iloc[i,j-1])+float(df.iloc[x,y]))/2
                elif j==0: #無效值發生在第一個column
                    if str.find(str(df.iloc[i,j+1]),'#')==-1 and str.find(str(df.iloc[i,j+1]),'*')==-1 and str.find(str(df.iloc[i,j+1]),'x')==-1 and str.find(str(df.iloc[i,j+1]),'A')==-1 and np.isnan(df.iloc[i,j+1])==False: 
                        df.iloc[i,j] = (float(df.iloc[i-18,23])+float(df.iloc[i,j+1]))/2
                    else:
                        while str.find(str(df.iloc[x,y]),'#')!=-1 or str.find(str(df.iloc[x,y]),'*')!=-1 or str.find(str(df.iloc[x,y]),'x')!=-1 or str.find(str(df.iloc[x,y]),'A')!=-1 or np.isnan(df.iloc[x,y])==True:
                            y+=1
                            if y==24:
                                x+=18
                                y = 0
                        df.iloc[i,j] = (float(df.iloc[i-18,23])+float(df.iloc[x,y]))/2
                elif j==23: #無效值發生在最後一column
                    while str.find(str(df.iloc[x+18,23-y]),'#')!=-1 or str.find(str(df.iloc[x+18,23-y]),'*')!=-1 or str.find(str(df.iloc[x+18,23-y]),'x')!=-1 or str.find(str(df.iloc[x+18,23-y]),'A')!=-1 or np.isnan(df.iloc[x+18,23-y])==True:
                        y-=1
                        if y==-1:
                            x+=18
                            y = 23
                    df.iloc[i,j] = (float(df.iloc[i,j-1])+float(df.iloc[x+18,23-y]))/2
            else:
                continue
end = time.time()
print(end-start)
```

```python
#確定無空值
df.info()
```

```python
#把日期 測站 測項組合回去
df = pd.concat([df_item,df],axis = 1)
```

```python
train = df.iloc[:1098,:] #訓練集 = 10.11月
test = df.iloc[1098:,:] #測試集 = 12月
test.index = range(0,len(test)) #訓練集從0開始編index
```

```python
train = train.drop(['日期','測站','測項'],axis = 1) 
test = test.drop(['日期','測站','測項'],axis = 1)
```

```python
#訓練集合併成(18,61*24)的DataFrame
train_split = train.iloc[:18,:] 
for i in range(18,1098,18):
    split = train.iloc[i:i+18,:]
    split.index = range(0,len(split))
    train_split = pd.concat([train_split,split],axis = 1,join_axes = [train_split.index])
```

```python
#測試集合併成(18,31*24)的DataFrame
test_split = test.iloc[:18,:]
for i in range(18,558,18):
    split = test.iloc[i:i+18,:]
    split.index = range(0,len(split))
    test_split = pd.concat([test_split,split],axis = 1,join_axes = [test_split.index])
```

```python
idx = ['00','01','02','03','04','05','06','07','08','09'] #訓練集columns名稱Rename成0~1463
for i in range(10,1464):
    idx.append(str(i))
train_split.columns = idx
```

```python
idx = ['00','01','02','03','44','05','06','07','08','09'] #測試集columns名稱Rename成0~743
for i in range(10,744):
    idx.append(str(i))
test_split.columns = idx
```

```python
train = train_split
test = test_split
```

```python
#import訓練的模型
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
#以只有PM2.5一種屬性預測
X = np.array(train)[9] #只有PM2.5的數值
X_1 = []
for i in range(0,len(X)-6): #0~5,1~6.......18~23依序取PM2.5的數值
    X_1.append(X[i:i+6])

Y_1 = []
for i in range(6,len(X)): #從第7小時開始取PM2.5的數值
    Y_1.append(X[i])

test_X = np.array(test)[9] #測試集的PM2.5數值
testX_1 = []
for i in range(0,len(test_X)-6): 
    testX_1.append(test_X[i:i+6])
test_Y = []
for i in range(6,len(test_X)): #測試集的7.8......23小時的PM2.5數值
    test_Y.append(test_X[i])
```

```python
test_X
```

```python
Y_1 #實際Target
```

```python
#只用PM2.5預測
regr_1 = LinearRegression() #線性回歸預測
regr_1.fit(X_1,Y_1) #擬合
y1_pred_lr = regr_1.predict(testX_1) #預測

print("以單一PM2.5線性回歸預測的MAE: ",mean_absolute_error(test_Y,y1_pred_lr))

rfr_1 = RandomForestRegressor(200) #隨機森林回歸預測
rfr_1.fit(X_1,Y_1)
y1_pred_rfr = rfr_1.predict(testX_1)

print("以單一PM2.5隨機森林回歸預測的MAE: ",mean_absolute_error(test_Y,y1_pred_rfr))
```

```python
#以18種屬性預測
#把訓練集每天18種屬性的所有數值合併成每天1筆
X = np.array(train)
X_18 = []
for i in range(0,len(X[0])-6): #0~6,1~6......18~23小時把18種屬性數值取出
    split_18 = []
    for j in range(0,len(X)):
        split_18.append(X[j][i:i+6])
    X_18.append(split_18)

t1 = []
new = []
for i in range(0,len(X_18)): #18種屬性0~5,1~6......18~23小時的數值每6個小時集合成1筆,每筆長度18*6=108
    for j in range(0,18):
        t2 = X_18[i][j]
        t1 = np.concatenate((t1,t2),axis = None)
    new.append(t1)
    t1 = []
X_18 = new
```

```python
X_18
```

```python
#把測試集每天18種屬性的所有數值合併成每天1筆
testX = np.array(test)
testX_18 = []
for i in range(0,len(testX[0])-6,1): #0~5,1~6......18~23小時把18種屬性數值取出
    split_18 = []
    for j in range(0,len(testX),1):
        split_18.append(testX[j][i:i+6])
    testX_18.append(split_18)

t1 = []
new = []
for i in range(0,len(testX_18)):#18種屬性0~5,1~6......18~23小時的數值每6個小時集合成1筆,每筆長度18*6=108
    for j in range(0,18):
        t2 = testX_18[i][j]
        t1 = np.concatenate((t1,t2),axis = None)
    new.append(t1)
    t1 = []
testX_18 = new
```

```python
testX_18
```

```python
#用18項屬性預測
regr_18 = LinearRegression() 
regr_18.fit(X_18,Y_1)
y18_pred_lr = regr_18.predict(testX_18)

print("18種屬性線性回歸預測的MAE: ",mean_absolute_error(test_Y,y18_pred_lr))

rfr_18 = RandomForestRegressor(200) 
rfr_18.fit(X_18,Y_1)
y18_pred_rfr = rfr_18.predict(testX_18)

print("18種屬性隨機森林回歸預測的MAE: ",mean_absolute_error(test_Y,y18_pred_rfr))
```
